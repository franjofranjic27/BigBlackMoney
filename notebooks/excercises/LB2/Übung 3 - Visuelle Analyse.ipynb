```python
### Daten einlesen
## Daten von Kaggle: https://www.kaggle.com/datasets/radheshyamkollipara/bank-customer-churn/data

# CSV-File einlesen - mit Library pandas
import pandas as pd
df = pd.read_csv('../../../data/Customer-Churn-Records.csv')
```

```python
# Erst einmal anschauen, was für Features vorhanden sind (Struktur des Datensatzes)

df.head()
```

```python
# Erstellen eines Boxplots (Simpel, mit pandas - bspw. für das Feature "EstimatedSalary")
df.boxplot(column=['EstimatedSalary'])
```

```python
# Erstellen eines Boxplots (etwas komplexer mit matplotlib - bspw. für Feature "EstimatedSalary")
# whis=1 definiert die Länge der "Whisker" als das 1.5-fache des IQA
# flierprops definiert das Aussehen der Ausreisser

import matplotlib.pyplot as plt

red_circle = dict(markerfacecolor='red', marker='o')
plt.boxplot(df['EstimatedSalary'], whis=0.5, flierprops=red_circle)
plt.show()
```

```python
# Erstellen eines Scatterplots (bspw. für die Features "Balance" und "EstimatedSalary")
plt.scatter(df['Balance'], df['EstimatedSalary'], alpha=0.3)
plt.xlabel('Balance')
plt.ylabel('EstimatedSalary')
plt.show()
```

```python
# Erstelle small multiples (bspw. für jede Region ein separater Scatterplot für "Alter" und "EstimatedSalary")
import matplotlib.pyplot as plt

# Erstelle ein Figure- und Axes-Objekt
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

# Definiere die Kategorien der dritten Variable
categories = df['Geography'].unique()

# Iteriere über die Kategorien und erstelle einen Scatterplot für jede Kategorie
for i, category in enumerate(categories):
  # Berechne die Position des Scatterplots in der Figure
  row = i // 2
  col = i % 2

  # Filtere die Daten für die aktuelle Kategorie
  data = df[df['Geography'] == category]

  # Erstelle den Scatterplot
  axes[row, col].scatter(data['Age'], data['EstimatedSalary'], alpha=0.3)

  # Füge Achsenbeschriftungen und Titel hinzu
  axes[row, col].set_xlabel("Alter")
  axes[row, col].set_ylabel("Geschätztes Gehalt")
  axes[row, col].set_title(f"Geographie: {category}")

# Passe den Abstand zwischen den Subplots an
plt.tight_layout()

# Zeige die Small Multiples an
plt.show()
```

```python
# Erstellen eines Histograms (bspw. für das Feature "Balance" mit 20 Balken)
plt.hist(df['Balance'], bins=20)
plt.xlabel('Balance')
plt.ylabel('Anzahl')
plt.show()
```

```python
# Erstelle wiederum ein Histogram und zeichne eine Standardverteilung ein (bspw. für das Feature "Balance")
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Berechne den Mittelwert und die Standardabweichung der Daten
mu, std = norm.fit(df['Balance'])

# Erstelle das Histogramm
plt.hist(df['Balance'], bins=20, density=True, alpha=0.6, color='g')

# Erstelle die x-Werte für die Standardverteilungskurve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)

# Berechne die y-Werte für die Standardverteilungskurve
p = norm.pdf(x, mu, std)

# Zeichne die Standardverteilungskurve
plt.plot(x, p, 'k', linewidth=2)

# Füge Titel und Achsenbeschriftungen hinzu
title = "Fit results: mu = %.2f,  std = %.2f" % (mu, std)
plt.title(title)
plt.xlabel('Balance')
plt.ylabel('Häufigkeit')

# Zeige das Histogramm mit der Standardverteilungskurve an
plt.show()
```
