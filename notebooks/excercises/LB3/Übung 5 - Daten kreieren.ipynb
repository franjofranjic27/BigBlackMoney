# Übung 5 - Daten kreieren

```python
### Daten einlesen
## Daten von Kaggle: https://www.kaggle.com/datasets/radheshyamkollipara/bank-customer-churn/data

# CSV-File einlesen - mit Library pandas
import pandas as pd
df = pd.read_csv('../../../data/Customer-Churn-Records.csv')
```

```python
df.head()
```

```python
# neues Feature "40p_100k" berechnen - bool'scher Wert für "Age" >40 und "EstimatedSalary" > 100'000
df['40p_100k'] = (df['Age'] > 40) & (df['EstimatedSalary'] > 100000)
df.head()
```
