# "Händische" komplette Berechnung

```python
import pandas as pd
from sympy import *

# Import data
points = list(pd.read_csv("https://bit.ly/2KF29Bd").itertuples())

display(points)
```

```python
# Definiere Funktionen und Variable
m, b, i, n = symbols('m b i n')
x, y = symbols('x y', cls=Function)

sum_of_squares = Sum((m*x(i) + b - y(i))**2, (i, 0, n))

d_m = diff(sum_of_squares, m) \
    .subs(n, len(points)-1).doit() \
    .replace(x, lambda i: points[i].x) \
    .replace(y, lambda i: points[i].y)

d_b = diff(sum_of_squares, b) \
    .subs(n, len(points)-1).doit() \
    .replace(x, lambda i: points[i].x) \
    .replace(y, lambda i: points[i].y)

d_m = lambdify([m,b], d_m)
d_b = lambdify([m,b], d_b)
```

```python
# Variablen initialisieren
m = 0.0
b = 0.0

L = 0.001

iterations = 100000
```

```python
# Berechnung durchführen und printen

for i in range(iterations):
  m_neu = m-d_m(m,b) * L
  b_neu = b-d_b(m,b) * L
  m = m_neu
  b = b_neu

print("y = {0}x + {1}".format(m,b))
```

# Einfache Lineare Regression mit Scikit Learn

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Import data
points = pd.read_csv("https://bit.ly/2KF29Bd")

x = points.values[:,0].reshape(-1, 1)
y = points.values[:,1].reshape(-1, 1)

model = LinearRegression()
model.fit(x,y)

print("y = {0}x + {1}".format(model.coef_[0],model.intercept_))
```

# Stochastischer Gradientenabstieg für lineare Regression

```python
# Standardimplementation der lineare Regression ist die geschlossene Form...
import pandas as pd
from sklearn.linear_model import LinearRegression

# Import data
points = pd.read_csv("https://bit.ly/2KF29Bd")

x = points.values[:,0].reshape(-1, 1)
y = points.values[:,1].reshape(-1, 1)

from sklearn.linear_model import SGDRegressor

# Lernrate ist 0.01, 1000 Iterationen
model = SGDRegressor(learning_rate='constant', eta0=0.01, max_iter=1000)
model.fit(x,y)

print("y = {0}x + {1}".format(model.coef_[0],model.intercept_))
```
